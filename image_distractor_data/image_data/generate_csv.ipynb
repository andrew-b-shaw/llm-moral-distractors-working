```python
import pandas as pd
import os
```

```python
dtypes = ['negative', 'neutral', 'positive']
file_paths = {}
for dtype in dtypes:
    files = [dtype + "/" + f for f in next(os.walk("./" + dtype))[2]]
    file_paths[dtype] = files
    df = pd.DataFrame(file_paths)
    pd.save_csv(df, f"../img_distractor_paths_{dtype}.csv", index=False)
```
