```python
import pandas as pd
import os
```

```python
sentiments = ['negative', 'neutral', 'positive']
modalities = ['img_distractor_data', 'text_distractor_data']
data = {'id': [], 'sentiment': [], 'file_path': [], 'modality': []}

id = 0
for m in modalities:
    for s in sentiments:
        files = [os.path.join(m, s, f) for f in next(os.walk(os.path.join(m, s)))[2]]
        for f in files:
            data['id'].append(id)
            data['sentiment'].append(s)
            data['file_path'].append(f)
            data['modality'].append(m)
            id += 1
df = pd.DataFrame(data)
df.head()

df.to_csv('distractors.csv', index=False)
```
